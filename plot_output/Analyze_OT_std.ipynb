```python
# this script tries to analyze the standard deviation of OT and surrouding cirrus BT
def read_text(fname):
    fo=open(fname,'r') #create file object
    #headstr=fo.readline()
    data=fo.readlines()
    #print (data)
    fo.close()  # close object
    return data

# main
import os 
import matplotlib as mpl
import matplotlib.pyplot as plt
#from mpl_toolkits.basemap import Basemap
import numpy as np
import pandas as pd
import glob

fnames=glob.glob('/data/keeling/a/yulanh/c/OT_output/*.txt')
#fnames=fnames[0:10]

OTA=np.zeros(1,'f')
OT_prob_trop=np.zeros(1,'f')
OT_prob_mid=np.zeros(1,'f')
OT_ave_bt_trop=np.zeros(1,'f')
Ci_ave_bt_trop=np.zeros(1,'f')
OT_ave_bt_mid=np.zeros(1,'f')
Ci_ave_bt_mid=np.zeros(1,'f')
OT_ave_std=np.zeros(1,'f')
Ci_ave_std=np.zeros(1,'f')
OTCi_ave_std_trop=np.zeros(1,'f')
OTCi_ave_std_mid=np.zeros(1,'f')

for fname in fnames:

    file_size=os.path.getsize(fname)
    
    if (file_size ==0):
        os.system('rm '+fname)
    
    if (file_size > 0):
        #print(fname)
        data=read_text(fname)
        splitcol=data[0].split(' ')
        Ncol=len(splitcol)-splitcol.count('')
        Nrow=len(data)
        #print(Ncol,Nrow)
        xscp=np.zeros(Nrow,'i')
        yscp=np.zeros(Nrow,'i')
        lon =np.zeros(Nrow,'f')
        lat =np.zeros(Nrow,'f')
        prob=np.zeros(Nrow,'f')
        ota =np.zeros(Nrow,'f')
        otbt=np.zeros(Nrow,'f')
        cibt=np.zeros(Nrow,'f')
        otstd=np.zeros(Nrow,'f')
        cistd=np.zeros(Nrow,'f')
        otcistd=np.zeros(Nrow,'f')
        
        # transfer data from string to float
        dataT=np.zeros((Nrow,Ncol),'f')
        for i in range(Nrow):
            splitcol=data[i].split(' ')
            k=0
            for j in range(len(splitcol)):
                if splitcol[j] != '' and splitcol[j] != '\n' : 
                    dataT[i,k]=float(splitcol[j])
                    k=k+1
            
            xscp[i]=dataT[i,0]
            yscp[i]=dataT[i,1]
            lon[i]=dataT[i,2]
            lat[i]=dataT[i,3]
            prob[i]=dataT[i,4]
            ota[i]=dataT[i,5]
            otbt[i]=dataT[i,7] # average otbt
            cibt[i]=dataT[i,11]
            otstd[i]=dataT[i,28]
            cistd[i]=dataT[i,29]
            otcistd[i]=dataT[i,30]
            
            #OT_prob=np.append(OT_prob,dataT[i,4])
            OTA=np.append(OTA,dataT[i,5])
            
            OT_ave_std=np.append(OT_ave_std,dataT[i,28])
            Ci_ave_std=np.append(Ci_ave_std,dataT[i,29])
            if (abs(lat[i]) < 30):
                OTCi_ave_std_trop=np.append(OTCi_ave_std_trop,dataT[i,30])
                OT_ave_bt_trop=np.append(OT_ave_bt_trop,dataT[i,7])
                Ci_ave_bt_trop=np.append(Ci_ave_bt_trop,dataT[i,11])
                OT_prob_trop=np.append(OT_prob_trop,dataT[i,4])
        
            if (abs(lat[i]) >=30):
                OTCi_ave_std_mid=np.append(OTCi_ave_std_mid,dataT[i,30])
                OT_ave_bt_mid=np.append(OT_ave_bt_mid,dataT[i,7])
                Ci_ave_bt_mid=np.append(Ci_ave_bt_mid,dataT[i,11])
                OT_prob_mid=np.append(OT_prob_mid,dataT[i,4])
        #print(otbt)
        #print(cibt)
        #print(otstd)
        #print(cistd)
        #print(otcistd)
        #print(OTCi_ave_std)


#plt.plot(OT_prob,OT_ave_bt,'bo')
print('finish')
```
