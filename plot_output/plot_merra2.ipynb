```python
from pyhdf.SD import SD, SDC
import h5py
import netCDF4 as nc
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.mpl.ticker as cticker
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

fontsz=14

fname='/data/keeling/a/yulanh/c/MERRA-2/2018/MERRA2_400.inst1_2d_asm_Nx.20181215.nc4'
ds = nc.Dataset(fname)
trop_T=ds['TROPT'][:]
lon   =ds['lon'][:]
lat   =ds['lat'][:]

fig,ax = plt.subplots(figsize=[16.5,8.5],subplot_kw={'projection': ccrs.PlateCarree()},constrained_layout=True)
im=ax.pcolormesh(lon,lat,trop_T[0,:,:],vmin=170,vmax=230)
p1=ax.scatter([164.8618],[42.5901],s=20,color='white')
ax.set_xlim([120,180])
ax.set_ylim([30,60])
ax.set_title('Mar. 15 2018 0 UTC')
ax.coastlines(resolution='110m', color='black', linewidth=1)
gl=ax.gridlines(draw_labels=True,linestyle='None')
gl.xlabels_top=False
gl.ylabels_right=False
gl.xlines=False
gl.ylines=False
gl.xlabel_style={'size':fontsz,'color':'black'}
gl.ylabel_style={'size':fontsz,'color':'black'}
cb=plt.colorbar(im,shrink=0.85,pad=0.04) 
cb.ax.tick_params(labelsize=fontsz)
cb.set_label(label='Tropopause Temperature (K)',size=fontsz)

fname='/data/keeling/a/yulanh/c/MERRA-2/2019/MERRA2_400.inst1_2d_asm_Nx.20190310.nc4'
ds = nc.Dataset(fname)
trop_T=ds['TROPT'][:]
lon   =ds['lon'][:]
lat   =ds['lat'][:]

fig,ax = plt.subplots(figsize=[16.5,8.5],subplot_kw={'projection': ccrs.PlateCarree()},constrained_layout=True)
im=ax.pcolormesh(lon,lat,trop_T[12,:,:],vmin=170,vmax=230)
p2=ax.scatter([-14.8047],[56.2979],s=20,color='white')
ax.set_xlim([-60,0])
ax.set_ylim([30,60])
ax.set_title('Mar. 10 2019 12 UTC')
ax.coastlines(resolution='110m', color='black', linewidth=1)
gl=ax.gridlines(draw_labels=True,linestyle='None')
gl.xlabels_top=False
gl.ylabels_right=False
gl.xlines=False
gl.ylines=False
gl.xlabel_style={'size':fontsz,'color':'black'}
gl.ylabel_style={'size':fontsz,'color':'black'}
cb=plt.colorbar(im,shrink=0.85,pad=0.04) 
cb.ax.tick_params(labelsize=fontsz)
cb.set_label(label='Tropopause Temperature (K)',size=fontsz)
```
