```python
#=== this script is to compare polygon and circular area

import matplotlib as mpl
import matplotlib.pyplot as plt
from datetime import datetime, date, timedelta
import numpy as np
import pandas as pd
from scipy import stats
import glob
from pyhdf.SD import SD, SDC
import h5py

years=['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013',\
       '2014','2015','2016','2017','2018','2019','2020']

nyear=len(years)
#nyear=1
NOTA=100 
Nhetero=60 # bin0.005, 0-0.2
NBTD_wv=100 # bin 0.5, 0-20 K
NBTD_ci=200 # bin 1, 0-60 K
NBTD_trop=200 # bin 0.5, 0-30K
NBT=90 #150-240
Nlat=180
NBT=90

All_OTA_pdf=np.zeros((Nlat,100,12,2),'i')
All_OTA_poly_pdf=np.zeros((Nlat,100,12,2),'i')

OT_area_10d=np.empty((0),'f') # obtain median area
OT_area_1025d=np.empty((0),'f')
OT_area_2540d=np.empty((0),'f')
OT_area_4060d=np.empty((0),'f')
OT_area_10n=np.empty((0),'f') # obtain median area
OT_area_1025n=np.empty((0),'f')
OT_area_2540n=np.empty((0),'f')
OT_area_4060n=np.empty((0),'f')

OT_poly_area_10d=np.empty((0),'f')
OT_poly_area_1025d=np.empty((0),'f')
OT_poly_area_2540d=np.empty((0),'f')
OT_poly_area_4060d=np.empty((0),'f')
OT_poly_area_10n=np.empty((0),'f')
OT_poly_area_1025n=np.empty((0),'f')
OT_poly_area_2540n=np.empty((0),'f')
OT_poly_area_4060n=np.empty((0),'f')

for year in years:  
    fname='/data/keeling/a/yulanh/c/OT_climatology/MODIS_OTNOT_summary_polygon2_'+year+'.h5'
    fid = h5py.File(fname,'r')
    OTA_pdf=fid['OTA_pdf']
    OTA_poly_pdf=fid['OTA_poly_pdf']
    a1=np.sum(OTA_pdf,axis=1)
    a2=np.sum(a1,axis=1)
    a3=np.sum(a2,axis=0)
    print(a3)
    OTA_day_10_points=fid['OTA_day_10_points']
    OTA_day_1025_points=fid['OTA_day_1025_points']
    OTA_day_2540_points=fid['OTA_day_2540_points']
    OTA_day_4060_points=fid['OTA_day_4060_points']
    OT_area_10d=np.append(OT_area_10d, OTA_day_10_points)
    #print('day',OT_area_10d.shape,OTA_day_10_points.shape)
    OT_area_1025d=np.append(OT_area_1025d, OTA_day_1025_points)
    OT_area_2540d=np.append(OT_area_2540d, OTA_day_2540_points)
    OT_area_4060d=np.append(OT_area_4060d, OTA_day_4060_points)
    
    OTA_night_10_points=fid['OTA_night_10_points']
    OTA_night_1025_points=fid['OTA_night_1025_points']
    OTA_night_2540_points=fid['OTA_night_2540_points']
    OTA_night_4060_points=fid['OTA_night_4060_points']
    OT_area_10n=np.append(OT_area_10n, OTA_night_10_points)
    #print('night',OT_area_10n.shape,OTA_night_10_points.shape)
    OT_area_1025n=np.append(OT_area_1025n, OTA_night_1025_points)
    OT_area_2540n=np.append(OT_area_2540n, OTA_night_2540_points)
    OT_area_4060n=np.append(OT_area_4060n, OTA_night_4060_points)
    
    OTA_poly_day_10_points=fid['OTA_poly_day_10_points']
    OTA_poly_day_1025_points=fid['OTA_poly_day_1025_points']
    OTA_poly_day_2540_points=fid['OTA_poly_day_2540_points']
    OTA_poly_day_4060_points=fid['OTA_poly_day_4060_points']
    OT_poly_area_10d=np.append(OT_poly_area_10d, OTA_poly_day_10_points)
    OT_poly_area_1025d=np.append(OT_poly_area_1025d, OTA_poly_day_1025_points)
    OT_poly_area_2540d=np.append(OT_poly_area_2540d, OTA_poly_day_2540_points)
    OT_poly_area_4060d=np.append(OT_poly_area_4060d, OTA_poly_day_4060_points)
    
    OTA_poly_night_10_points=fid['OTA_poly_night_10_points']
    OTA_poly_night_1025_points=fid['OTA_poly_night_1025_points']
    OTA_poly_night_2540_points=fid['OTA_poly_night_2540_points']
    OTA_poly_night_4060_points=fid['OTA_poly_night_4060_points']
    OT_poly_area_10n=np.append(OT_poly_area_10n, OTA_poly_night_10_points)
    OT_poly_area_1025n=np.append(OT_poly_area_1025n, OTA_poly_night_1025_points)
    OT_poly_area_2540n=np.append(OT_poly_area_2540n, OTA_poly_night_2540_points)
    OT_poly_area_4060n=np.append(OT_poly_area_4060n, OTA_poly_night_4060_points)
    
```

```python
#combine all latitudes
OTA_circular_day=np.empty((0),'f')
OTA_circular_day=np.append(OTA_circular_day,OT_area_10d)
#print(OTA_circular_day.shape,OT_area_10d.shape)
OTA_circular_day=np.append(OTA_circular_day,OT_area_1025d)
#print(OTA_circular_day.shape,OT_area_1025d.shape)
OTA_circular_day=np.append(OTA_circular_day,OT_area_2540d)
#print(OTA_circular_day.shape,OT_area_2540d.shape)
OTA_circular_day=np.append(OTA_circular_day,OT_area_4060d)
#print(OTA_circular_day.shape,OT_area_4060d.shape)

OTA_circular_night=np.empty((0),'f')
OTA_circular_night=np.append(OTA_circular_night,OT_area_10n)
print(OTA_circular_night.shape,OT_area_10n.shape)
OTA_circular_night=np.append(OTA_circular_night,OT_area_1025n)
print(OTA_circular_night.shape,OT_area_1025n.shape)
OTA_circular_night=np.append(OTA_circular_night,OT_area_2540n)
print(OTA_circular_night.shape,OT_area_2540n.shape)
OTA_circular_night=np.append(OTA_circular_night,OT_area_4060n)
print(OTA_circular_night.shape,OT_area_4060n.shape)

OTA_poly_day=np.empty((0),'f')
OTA_poly_day=np.append(OTA_poly_day,OT_poly_area_10d)
OTA_poly_day=np.append(OTA_poly_day,OT_poly_area_1025d)
OTA_poly_day=np.append(OTA_poly_day,OT_poly_area_2540d)
OTA_poly_day=np.append(OTA_poly_day,OT_poly_area_4060d)

OTA_poly_night=np.empty((0),'f')
OTA_poly_night=np.append(OTA_poly_night,OT_poly_area_10n)
OTA_poly_night=np.append(OTA_poly_night,OT_poly_area_1025n)
OTA_poly_night=np.append(OTA_poly_night,OT_poly_area_2540n)
OTA_poly_night=np.append(OTA_poly_night,OT_poly_area_4060n)
#print(OTA_poly_day.shape,OTA_poly_night.shape)

ind=np.where(OTA_poly_day < 1000)[0]
OTA_poly_day1=OTA_poly_day[ind]
OTA_circ_day1=OTA_circular_day[ind]
# derive histogram
poly_his=np.histogram(OTA_poly_day1,bins=200,range=(0,1000))
poly_day_his=poly_his[0]
circ_his=np.histogram(OTA_circ_day1,bins=200,range=(0,1000))
circ_day_his=circ_his[0]
ind=np.where(OTA_poly_night < 1000)[0]
OTA_poly_night1=OTA_poly_night[ind]
OTA_circ_night1=OTA_circular_night[ind]
# derive histogram
poly_his=np.histogram(OTA_poly_night1,bins=200,range=(0,1000))
poly_night_his=poly_his[0]
circ_his=np.histogram(OTA_circ_night1,bins=200,range=(0,1000))
circ_night_his=circ_his[0]

#== to plot ===
nbins=201
x=circ_his[1]
x=x[1:nbins+1]
linewidth=2
fontsz=15
fig,ax =plt.subplots(1,figsize=([5.5,5]))
p0=ax.plot(x,poly_day_his/len(OTA_poly_day1),color='black',linewidth=linewidth)
p1=ax.plot(x,circ_day_his/len(OTA_circ_day1),color='black',linestyle='dashed',linewidth=linewidth)
p2=ax.plot(x,poly_night_his/len(OTA_poly_night1),color='gray',linewidth=linewidth)
p3=ax.plot(x,circ_night_his/len(OTA_circ_night1),color='gray',linestyle='dashed',linewidth=linewidth)
ax.set_xlim([0,300])
ax.set_ylim([0,0.13])
ax.set_xlabel('OTA ($km^2$)',size=fontsz)
ax.set_ylabel('Frequency',size=fontsz)
#ax.set_xticks([0,20,40,60,80,100,120,140,160,180,200])
ax.tick_params(labelsize=fontsz)
ax.legend(['Polygon area, day','Circular area, day','Polygon area, night','Circular area, night'],fontsize=12)
#print(poly_day_his/len(OTA_poly_day1))
#print(poly_night_his/len(OTA_poly_night1))
```
