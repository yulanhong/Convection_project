```python
# main
import os 
import matplotlib as mpl
import matplotlib.pyplot as plt
#from mpl_toolkits.basemap import Basemap
import numpy as np
import pandas as pd
import glob
from pyhdf.SD import SD, SDC
import h5py
import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker
from datetime import datetime, date, timedelta

print('start',datetime.now())

year='2021'

fnames=sorted(glob.glob('/data/keeling/a/yulanh/f/OT_project/Aqua/*'+year+'*'))


latres=1
lonres=1
xdim=int(360/lonres)
ydim=int(180/latres)
zdim=12

all_obsnum=np.zeros((ydim,xdim,zdim),'i')
all_num_bt180 = np.zeros((ydim,xdim,zdim),'i')
all_num_bt200 = np.zeros((ydim,xdim,zdim),'i')
all_num_bt210 = np.zeros((ydim,xdim,zdim),'i')
all_num_bt220 = np.zeros((ydim,xdim,zdim),'i')
all_num_bt230 = np.zeros((ydim,xdim,zdim),'i')
all_num_bt240 = np.zeros((ydim,xdim,zdim),'i')
all_num_cdtp = np.zeros((ydim,xdim,zdim),'i')
all_num_cdtp2 = np.zeros((ydim,xdim,zdim),'i')
all_num_cdtp5 = np.zeros((ydim,xdim,zdim),'i')
all_num_cdtp10 = np.zeros((ydim,xdim,zdim),'i')

for fname in fnames:

    Julday=int(fname[-10:-7])
    start_date=date(int(year),1,1)
    convt_date=start_date+timedelta(days=int(Julday)-1)
    date_res  = convt_date.strftime("%m-%d-%y")
    month_st=date_res[0:2]
    month_scp=int(month_st)-1
    #print(fname,month_scp,Julday)

    hdf=SD(fname)
    obsnum=hdf.select('obs_num')
    obsnum=obsnum[:,:]
    all_obsnum[:,:,month_scp]=all_obsnum[:,:,month_scp]+obsnum
    
    num_bt180=hdf.select('num_bt180')
    num_bt180=num_bt180[:,:]
    all_num_bt180[:,:,month_scp]=all_num_bt180[:,:,month_scp] + num_bt180
    
    num_bt200=hdf.select('num_bt200')[:,:]
    all_num_bt200[:,:,month_scp]=all_num_bt200[:,:,month_scp] + num_bt200
    
    num_bt210=hdf.select('num_bt210')[:,:]
    all_num_bt210[:,:,month_scp]=all_num_bt210[:,:,month_scp] + num_bt210
    
    num_bt220=hdf.select('num_bt220')[:,:]
    all_num_bt220[:,:,month_scp]=all_num_bt220[:,:,month_scp] + num_bt220
    
    num_bt230=hdf.select('num_bt230')[:,:]
    all_num_bt230[:,:,month_scp]=all_num_bt230[:,:,month_scp] + num_bt230
    
    num_bt240=hdf.select('num_bt240')[:,:]
    all_num_bt240[:,:,month_scp]=all_num_bt240[:,:,month_scp] + num_bt240
    
    num_cdtp=hdf.select('num_colder_tropopause')[:,:]
    all_num_cdtp[:,:,month_scp]=all_num_cdtp[:,:,month_scp] + num_cdtp
    
    num_cdtp2=hdf.select('num_colder_tropopause2')[:,:]
    all_num_cdtp2[:,:,month_scp]=all_num_cdtp2[:,:,month_scp] + num_cdtp2
    
    num_cdtp5=hdf.select('num_colder_tropopause5')[:,:]
    all_num_cdtp5[:,:,month_scp]=all_num_cdtp5[:,:,month_scp] + num_cdtp5
    
    num_cdtp10=hdf.select('num_colder_tropopause10')[:,:]
    all_num_cdtp10[:,:,month_scp]=all_num_cdtp10[:,:,month_scp] + num_cdtp10

# write the monthly data

wfname='/data/keeling/a/yulanh/f/OT_project/Aqua/monthly/Aqua_obsnum_202101.hdf'
hf=h5py.File(wfname,'w')
hf.create_dataset('obs_num',data=all_obsnum)
hf.create_dataset('num_bt180',data=all_num_bt180)
hf.create_dataset('num_bt200',data=all_num_bt200)
hf.create_dataset('num_bt210',data=all_num_bt210)
hf.create_dataset('num_bt220',data=all_num_bt220)
hf.create_dataset('num_bt230',data=all_num_bt230)
hf.create_dataset('num_bt240',data=all_num_bt240)
hf.create_dataset('num_cdtp',data=all_num_cdtp)
hf.create_dataset('num_cdtp2',data=all_num_cdtp2)
hf.create_dataset('num_cdtp5',data=all_num_cdtp5)
hf.create_dataset('num_cdtp10',data=all_num_cdtp10)
hf.close()

print('finish',datetime.now())
```
